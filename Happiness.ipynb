# Removing Outliers

```python
columns = ['Standard Error', 'Family',
        'Trust (Government Corruption)',
       'Generosity', 'Dystopia Residual']
for col in columns:
    IQR = df[col].quantile(0.75)-df[col].quantile(0.25)
    Barmax =  df[col].quantile(0.75) + 1.5*IQR
    Barmin =  df[col].quantile(0.25) - 1.5*IQR
    df.loc[df[col]>Barmax,col] = Barmax
    df.loc[df[col]<Barmin,col] = Barmin
```

Outliers are replaced in the dataset

# Labeling the dataset

```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
for col in ['Country','Region']:
    df[col] =  LE.fit_transform(df[col])
```

Converted Object type to int using Labels for categorical data

# Scaling the data

```python
from sklearn import preprocessing
scaler = preprocessing.MinMaxScaler()
minmax_df = scaler.fit_transform(df)
final_df = pd.DataFrame(minmax_df, columns = df.columns)
```

Data is scaled now

# Removing the skewness in data

```python
fin_df = x = final_df.drop(["Country","Region"],axis=1)
from sklearn.preprocessing import power_transform
fin_df_new=power_transform(fin_df)
fin_df_new = pd.DataFrame(fin_df_new,columns=fin_df.columns)
```

```python
fin_df_new = fin_df_new.join(df['Country'])
fin_df_new = fin_df_new.join(df['Region'])
```

Skewness in data set is removed

# Train_Test_Split

```python
x = fin_df_new.drop(['Happiness Score'],axis=1)
y = fin_df_new["Happiness Score"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=.30,random_state=5)
```

# Model Building

-----------Linear Regression-----------

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error
regr = LinearRegression()
regr.fit(x_train, y_train)
print(regr.score(x_test, y_test))
```

```python
regr_pred = regr.predict(x_test)
regr_mse = mean_squared_error(y_test,regr_pred)
rmse = np.sqrt(regr_mse)
print(rmse)
```

```python
print(r2_score(y_test, regr_pred))
```

Linear Regression:

    Score : 0.99
    MSE   : 0.075
    R2 Score: 0.99

-----------Decision tree Regression-----------

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
DTRM = DecisionTreeRegressor()
DTRM.fit(x_train,y_train)
```

```python
DTRM_pred = DTRM.predict(x_test)
```

```python
mse = mean_squared_error(y_test,DTRM_pred)
rmse = np.sqrt(mse)
rmse
```

```python
print(r2_score(y_test, DTRM_pred))
```

Decision Tree Regression:

    MSE   : 0.08
    R2 Score: 0.992

-----------Ridge Regression-----------

```python
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
```

```python
rr = Ridge(alpha=0.01)
rr.fit(x_train, y_train) 
predict= rr.predict(x_train)
print(np.sqrt(mean_squared_error(y_train,predict)))
print(r2_score(y_train, predict))
```

Ridge Regression:

    MSE   : 0.11
    R2 Score: 0.98

-----------Lasso Regression-----------

```python
model_lasso = Lasso(alpha=0.01)
model_lasso.fit(x_train, y_train) 
pred_train_lasso= model_lasso.predict(x_train)
print(np.sqrt(mean_squared_error(y_train,pred_train_lasso)))
print(r2_score(y_train, pred_train_lasso))
```

Lasso Regression:

    MSE   : 0.12
    R2 Score: 0.98

-----------Elastic Net Regression-----------

```python
model_enet = ElasticNet(alpha = 0.01)
model_enet.fit(x_train, y_train) 
pred_train_enet= model_enet.predict(x_train)
print(np.sqrt(mean_squared_error(y_train,pred_train_enet)))
print(r2_score(y_train, pred_train_enet))
```

Lasso Regression:

    MSE   : 0.11
